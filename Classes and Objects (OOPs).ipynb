## Class and Object. 

Class : Class is a blue print of an Object, class is actually tells about the behavior of an object.

Object : Object is an instance of a class 

Python has two types of class built-in class (like list, tuple, set, dictionary) and user defined class.

```python
class Atm(): 

    # constructor is a sepcial function, that will instantiate when object is created. 
    def __init__(self): 
        self.pin = ''
        self.balance = 0
        print('Object is Created.')
        self.menu()

    def menu(self): 
        user_input = input("""
                Hi, how can i help you ? 
                1. Press 1 to create pin 
                2. Press 2 to change pin 
                3. Press 3 to check balance 
                4. Press 4 to withdraw
                5. Press anything to exit
        """)

        if user_input == '1': 
            # create a pin.
            self.create_pin()
        elif user_input == '2': 
            # change the pin 
            self.change_pin()
        elif user_input == '3':
            # check balance
            self.check_balance()
        elif user_input == '4': 
            # withdraw amount 
            self.withdraw()
        if user_input == '5': 
            exit()

    def create_pin(self): 
        initial_pin = input('\t\tEnter the pin : ')
        self.pin = initial_pin

        intial_balance = int(input('\t\tEnter your Balance : '))
        self.balance = intial_balance
        
        print('\t\tPin Created Successfully.')

        self.menu()

    
    def change_pin(self): 
        old_pin = input('\t\tEnter your Last Pin : ')
        if old_pin == self.pin: 
            # change the pin 
            new_pin = input('\t\tEnter the New pin : ')
            self.pin = new_pin
            print('\t\tPin changed Successfully')
            self.menu()
        else: 
            # Not allowed to change the pin 
            print('\t\tSorry, You Entered the wroung Pin')
            self.menu()


    def check_balance(self): 
        user_pin = input('\t\tEnter the pin : ')
        if user_pin == self.pin: 
            print('\t\tYour balance is : ', self.balance)
        else: 
            print('\t\tSoory, Enter the correct Pin')

        self.menu()

    def withdraw(self): 
        user_pin = input('\t\tEnter the Pin : ')
        if user_pin == self.pin: 
            # withdrawl is done.
            withdrawal_amount = int(input('\t\tEnter the amount you want to withdrwal : '))
            if withdrawal_amount <= self.balance: 
                self.balance = self.balance - withdrawal_amount
                print('\t\tAmount withdrawal Successfully')
                print('\t\tYour Current Balance is ', self.balance)
            else: 
                print('\t\tInsufficient Funds.')
        else: 
            print('\t\tSorry, Enter the correct Pin')   
        self.menu()
```

## Create the Own DataType called Fraction.  
- Here we are using many magic function
- We write some comment about when these function are triggered.

```python
class Fraction: 
    
    # Parametrized Constructor of this class. 
    def __init__(self, num1, num2): 
        self.num = num1
        self.den = num2

    # This is another magic function that will called when we trying to print the object of the class using print function.
    def __str__(self): 
        return '{}/{}'.format(self.num, self.den)

    # another magic function that will trigger when we do add operation of two objects using plus(+) operator. 
    def __add__(self, other): 
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den

        return '{}/{}'.format(new_num, new_den)

    # another magic function that will trigger when we do subtraction operation of two objects using minus(-) operator. 
    def __sub__(self, other): 
        new_num = self.num * other.den - self.den * other.num
        new_den = self.den * other.den

        return '{}/{}'.format(new_num, new_den)

    # another magic function that will trigger when we do Multiplication operation of two objects using Asterik(*) operator. 
    def __mul__(self, other): 
        new_num = self.num * other.num
        new_den = self.den * other.den

        return '{}/{}'.format(new_num, new_den)

    # another magic function that will trigger when we do Division operation of two objects using slash(/) operator. 
    def __truediv__(self, other): 
        new_num = self.num * other.den
        new_den = self.den * other.num

        return '{}/{}'.format(new_num, new_den)
```

```python
fr1 = Fraction(2, 3)
fr2 = Fraction(4, 5)
```

```python
print(fr1)
print(fr2)
```

```python
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```
