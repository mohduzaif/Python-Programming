### All the NLP task are doing by using the APIs provided by NLP Cloud

```python
import nlpcloud

class NLPTask: 

    def __init__(self): 
        self.__database = {} 
        self.__first_menu()

        
    def __first_menu(self):
        first_input = input(""" 
        Hi! how would you like to proceed?
        1. Not a member? Register
        2. Already a member? Login
        3. For Exit? Exit
        """)

        if first_input == '1': 
            self.__register()
        elif first_input == '2': 
            self.__login()
        else: 
            exit() 


    def __second_menu(self):
        second_input = input("""
            Hi! how would you like to proceed?
            1. NER
            2. Language Detection
            3. Sentiment Analysis
            4. Sementic Similarity
            5. code Generation
            6. Logout
            """)

        if second_input == '1': 
            self.__ner()
        elif second_input == '2': 
            self.__language_detection()
        elif second_input == '3':
            self.__sentiment_analysis()
        elif second_input == '4': 
            self.__sementic_similarity()
        elif second_input == '5': 
            self.__code_generation()
        else: 
            exit()

    
    # Register method. 
    def __register(self): 
        name = input('Enter name : ')
        email = input('Enter Email : ')
        pswd = input('Enter password : ')

        if email in self.__database: 
            print('Already Registered')
            self.__login()
        else: 
            self.__database[email] = [name, pswd]
            print('Registration Successful')
            # print(self.__database)
            self.__first_menu()


    # Login Method
    def __login(self):
        email = input('Enter Email : ')
        pswd = input('Enter Password : ')

        if email in self.__database: 
            if self.__database[email][1] == pswd: 
                print('Login Successfully')
                self.__second_menu()
            else: 
                print('You Entered the Wrong Password')
                self.__login()
        else: 
            print('Email is not registered, Firstly Registered')
            self.__first_menu()


    # It solves a very interesting challenge: extracting entities from a block of text.
    def __ner(self): 
        text = input('Enter the Text : ')
        entity_to_search = input('Enter the text you want to Search : ')
        
        client = nlpcloud.Client("finetuned-llama-3-70b", "d4ac68d1607984d2fe23c20157d68cacb1ccba23", gpu=True)
        response = client.entities(text, searched_entity=entity_to_search)

        # print(response)
        counter = 0
        for language in response['entities']: 
            counter += 1
            print(language['text'])
        print('Total Count is ', counter)

        self.__second_menu()


    # Detect the language from the text. 
    def __language_detection(self): 
        text = input('Enter the Text : ')
        
        client = nlpcloud.Client("python-langdetect", "d4ac68d1607984d2fe23c20157d68cacb1ccba23", gpu=False)
        response = client.lang_detection(text)

        print(response)

        self.__second_menu()


    # Do the sentiment Analysis from the text we provided in the input.
    def __sentiment_analysis(self):
        text = input('Enter the Text : ')

        client = nlpcloud.Client("finetuned-llama-3-70b", "d4ac68d1607984d2fe23c20157d68cacb1ccba23", gpu=True)
        response = client.sentiment(text, target="NLP Cloud")

        print(response)

        self.__second_menu()


    # Sementic similarity between two texts. 
    def sementic_similarity(self): 

        text1 = input('Enter the first Text : ')
        text2 = input('Enter the second Text : ')
        
        client = nlpcloud.Client("paraphrase-multilingual-mpnet-base-v2", "d4ac68d1607984d2fe23c20157d68cacb1ccba23", gpu=False)
        response = client.semantic_similarity([text1, text2])
        
        # print(response)

        L = []
        for i in response['scored_labels']:
            L.append(i['score'])

        index = sorted(list(enumerate(L)),key=lambda x:x[1],reverse=True)[0][0]
        print(response['scored_labels'][index]['label'])

        self.__second_menu()

    
    # Code geration method. 
    def __code_generation(self): 
        problem = input('Enter your problem in details : ')
        
        client = nlpcloud.Client("finetuned-llama-3-70b", "d4ac68d1607984d2fe23c20157d68cacb1ccba23", gpu=True)
        code = client.code_generation(problem)

        print(code)

        self.__second_menu()
```
