#### Problem 1 - Print the following pattern. Write a program to use for loop to print the following reverse number pattern.

```python
# 5 4 3 2 1 
# 4 3 2 1 
# 3 2 1 
# 2 1 
# 1

n = int(input('Enter the value of n : '))

for i in range(n, 0, -1): 
    for j in range(i, 0, -1): 
        print(j, end=' ')
    print()
```

#### Problem 6: The natural logarithm can be approximated by the following series.


If x is input through the keyboard, write a program to calculate the sum of the first seven terms of this series.

```python
x = int(input('Enter the value of x : ')) 
n = 7

sum_of_sequence = 0
for i in range(1, n + 1):
    sum_of_sequence += (1 / i) * ((x - 1) / x)**i

print(sum_of_sequence)
```

#### Problem 9: Write a program that will take a decimal number as input and prints out the binary equivalent of the number.

```python
n = int(input('Enter the number : '))

result = ''

while n != 0: 
    rem = n % 2
    n = n // 2
    result = str(rem) + result

print(result)
print(type(result))
```

#### `Problem 10` : Write a program that will take 2 numbers as input and prints the LCM and HCF of those 2 numbers

```python
number1 = int(input('Enter the first number : ')) 
number2 = int(input('Enter the second number : '))

number1 = max(number1, number2)
number2 = min(number1, number2)

# find the HCF of the given two numbers. 
while True: 
    remainder = number1 % number2
    if remainder == 0: 
        print('HCF of given two numbers is : ', number2)
        break
    number1 = number2
    number2 = remainder


# find the LCM of the given two numbers. 
i = number1
while True:
    if i % number2 == 0:
        print('LCM is : ', i) 
        break
    i += number1
```

#### `Problem 11 :` Create Short Form from initial character. 

Given a string create short form of the string from Initial character. Short form should be capitalised.

```python
# Example : Data science mentorship program   ---->  DSMP

s = input('Enter the String : ')

list = []
list.append(s[0])

for index in range(0, len(s)-1): 
    if s[index] == ' ':
        list.append(s[index + 1].upper())

print(''.join(list))
```

#### `Problem 13 :` Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.

```python
# str1 = PyNaTive
# OUTPUT : yaivePNT

str1 = input('Enter the String : ')

lower_case = ''
upper_case = ''

result = ''

for ch in str1: 
    if ch >= 'A' and ch <= 'Z': 
        upper_case += ch 
    else: 
        lower_case += ch 
    
result = lower_case + upper_case
print(result)
```

#### `Problem 14 :` Take a alphanumeric string input and print the sum and average of the digits that appear in the string, ignoring all other characters.

```python
# INPUT : hel123O4every093 

str1 = input('Enter the String : ')

digit = ''
for ch in str1: 
    if ch >= '0' and ch <= '9': 
        digit += ch 

print(digit)

sum = 0
for ch in digit: 
    sum += int(ch) 
print('Sum is : ', sum)
print('Average is : ', sum / len(digit))
```
